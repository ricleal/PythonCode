```python
%matplotlib inline
#%matplotlib notebook

import matplotlib.pyplot as plt

plt.rcParams['image.cmap'] = 'viridis'
plt.rcParams['figure.figsize'] = 8*1.2, 6*1.2

import xarray as xr
import numpy as np
```

```python
# Generate dumy data
x = np.linspace(-75, 75, 256)
y = np.linspace(-100, 100, 256)
xv, yv = np.meshgrid(x, x)
r = np.sqrt((xv+75)**2 + (yv)**2)

# create our 2d array
data = xr.DataArray(
    r,
    [('y', y),
     ('x', x)]
)

data
```

```python
data.plot()
```

```python
# let's get the 2D data as numpy array and plot it
import matplotlib.pyplot as plt
z = data.values
plt.imshow(z)
```

```python
# Let's add sum metadata
data.attrs
```

```python
data.attrs['x_min'] = -50
data.attrs['x_max'] = 50

data_within_limits = data[(data.x > data.attrs['x_min']) & (data.x < data.attrs['x_max'])]
data_within_limits
```

```python
data.plot()
```
