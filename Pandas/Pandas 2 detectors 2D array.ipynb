# Converts 2D numpy array into pandas Dataframe and back into a 2D array
## 2 detec9


Matrix ij convention
i is Row
j is column

2 detectors

`name,i,j,x,y,z,values,errors`

```python
%matplotlib notebook
#%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import OrderedDict

import matplotlib.pyplot as plt
plt.rcParams['image.cmap'] = 'viridis'
```

```python
# Det1
#
# Shape of detector
n_rows, n_cols = shape = (2, 5)  # pos 0 rows, pos 1 columns
total_size = n_rows * n_cols

value = np.arange(total_size).reshape(n_rows, n_cols)
error = np.sqrt(value)
rows_v, cols_v = np.meshgrid(range(n_rows), range(n_cols), indexing='ij')
# Simulate x,y,values
x_v, y_v = np.meshgrid(np.linspace(1,2,n_rows), np.linspace(1,-1,n_cols), indexing='ij')

det1 = {'name': np.full(total_size, 'det1', dtype=np.dtype('S4')),
     'i': rows_v.ravel(),  # i = rows
     'j': cols_v.ravel(),  # j = coluns
     'x': x_v.ravel(),
     'y': y_v.ravel(),
     'value': value.ravel(),
     'error': error.ravel(),
     }

df1 = pd.DataFrame(det1)
# Reorder columns
df1 = df1[['name', 'i', 'j', 'x', 'y', 'value', 'error']]
df1
```

```python
#
# Det2
#

# Shape of detectors
n_rows, n_cols = shape = (3, 4)  # pos 0 rows, pos 1 columns
total_size = n_rows * n_cols

value = np.arange(total_size).reshape(n_rows, n_cols)
error = np.sqrt(value)
rows_v, cols_v = np.meshgrid(range(n_rows), range(n_cols), indexing='ij')
# Simulate x,y,values
x_v, y_v = np.meshgrid(np.linspace(2,4,n_rows), np.linspace(1,-1,n_cols), indexing='ij')

det2 = {'name': np.full(total_size, 'det2', dtype=np.dtype('S4')),
     'i': rows_v.ravel(),  # i = rows
     'j': cols_v.ravel(),  # j = coluns
     'x': x_v.ravel(),
     'y': y_v.ravel(),
     'value': value.ravel(),
     'error': error.ravel(),
     }


df2 = pd.DataFrame(det2)
# Reorder columns
df2 = df2[['name', 'i', 'j', 'x', 'y', 'value', 'error']]
df2
```

```python
df = df1.append(df2, ignore_index=True)
print("---- DataFrame ----")
df
```

```python
pivot1 = df[df.name == "det1".encode('utf-8')].pivot(index='i', columns='j', values='value')
print("---- DataFrame Pivot Det1 ----")
pivot1
```

```python
plt.figure()
plt.imshow(pivot1.values)
plt.colorbar()
```

```python
from importlib import import_module


class Data(object):
    '''
    Represents the data frame as a class for the users
    '''
    def __init__(self,df):
        names = df.name.unique()
        for name_binary in names:
            name = name_binary.decode('utf-8')
            setattr(self,name,lambda: None)
            setattr(self.__dict__[name],"value",df[df.name == name_binary].pivot(index='i', columns='j', values='value'))
            setattr(self.__dict__[name],"error",df[df.name == name_binary].pivot(index='i', columns='j', values='error'))

d = Data(df)
print(d.det1.value)
print(d.det1.error)
print(d.det2.value)
print(d.det2.error)

```
