# Converts 2D numpy array into pandas Dataframe and back into a 2D array
## 2 detec9


Matrix ij convention
i is Row
j is column

2 detectors

`name,i,j,x,y,z,values,errors`

```python
%matplotlib notebook
#%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import OrderedDict

import matplotlib.pyplot as plt
plt.rcParams['image.cmap'] = 'viridis'
```

```python
# Det1
#
# Shape of detector
n_rows, n_cols = shape = (2, 5)  # pos 0 rows, pos 1 columns
total_size = n_rows * n_cols

value = np.arange(total_size).reshape(n_rows, n_cols)
error = np.sqrt(value)
rows_v, cols_v = np.meshgrid(range(n_rows), range(n_cols), indexing='ij')
# Simulate x,y,values
x_v, y_v = np.meshgrid(np.linspace(1,2,n_rows), np.linspace(1,-1,n_cols), indexing='ij')

det1 = OrderedDict({'name': np.full(total_size, 'det1', dtype=np.dtype('S4')),
     'i': rows_v.ravel(),  # i = rows
     'j': cols_v.ravel(),  # j = coluns
     'x': x_v.ravel(),
     'y': y_v.ravel(),
     'value': value.ravel(),
     'error': error.ravel(),
     })

df1 = pd.DataFrame(det1)
df1
```

```python
#
# Det2
#

# Shape of detectors
n_rows, n_cols = shape = (3, 4)  # pos 0 rows, pos 1 columns
total_size = n_rows * n_cols

value = np.arange(total_size).reshape(n_rows, n_cols)
error = np.sqrt(value)
rows_v, cols_v = np.meshgrid(range(n_rows), range(n_cols), indexing='ij')
# Simulate x,y,values
x_v, y_v = np.meshgrid(np.linspace(2,4,n_rows), np.linspace(1,-1,n_cols), indexing='ij')

det2 = OrderedDict({'name': np.full(total_size, 'det2', dtype=np.dtype('S4')),
     'i': rows_v.ravel(),  # i = rows
     'j': cols_v.ravel(),  # j = coluns
     'x': x_v.ravel(),
     'y': y_v.ravel(),
     'value': value.ravel(),
     'error': error.ravel(),
     })


df2 = pd.DataFrame(det2)
df2
```

```python
df = df1.append(df2, ignore_index=True)
print("---- DataFrame ----")
df
```

```python
pivot = df[df.name == "det1".encode('utf-8')].pivot(index='i', columns='j', values='value')
print("---- DataFrame Pivot Det1 ----")
pivot
```

```python
plt.figure()
plt.imshow(pivot.values)
plt.colorbar()
```

```python
pivot = df[df.name == "det2".encode('utf-8')].pivot(index='i', columns='j', values='value')
print("---- DataFrame Pivot Det2 ----")
pivot
```

```python
plt.figure()
plt.imshow(pivot.values)
plt.colorbar()
```

```python
# Let's add some axis
x = df[df.name == "det2".encode('utf-8')].x.unique()
y = df[df.name == "det2".encode('utf-8')].y.unique()
print("x = {}".format(x))
print("y = {}".format(y))
print(pivot.values)

plt.figure()
plt.imshow(pivot.values,extent=(x.min(), x.max(), y.min(), y.max()),origin='upper')
plt.colorbar()
```

```python
# pcolormesh chops last collumns!!
print(pivot.values)
x = df[df.name == "det2".encode('utf-8')].x
y = df[df.name == "det2".encode('utf-8')].y
plt.figure()
plt.pcolormesh(x.reshape(shape), 
               y.reshape(shape), 
               pivot.values)
plt.colorbar()
```
