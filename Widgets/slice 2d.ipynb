```python
%matplotlib inline

from IPython.html.widgets import interact
import matplotlib.pyplot as plt
import numpy as np

N = 20
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
data = np.random.rand(300,300,N)

def plot_2d(n):
    plt.imshow(data[:,:,n])
```

```python
interact(plot_2d, n=(0,N-1, 1),)
```

```python
x, y, z = np.ogrid[-10:10:20j, -10:10:20j, -10:10:20j]
s = np.sin(x*y*z)/(x*y*z)
```

```python
data = s
def plot_slice(n,dim):
    if dim==1:
        plt.title('Varying X')
        plt.imshow(data[n,:,:])
        plt.xlabel('y')
        plt.ylabel('z')
    elif dim==2:
        plt.title('Varying Y')
        plt.imshow(data[:,n,:])
        plt.xlabel('x')
        plt.ylabel('z')
    else:
        plt.title('Varying Z')
        plt.imshow(data[:,:,n])
        plt.xlabel('x')
        plt.ylabel('y')
        
interact(plot_slice, n=(0,19, 1),dim=(1,3, 1))
```

```python
from IPython.display import display
from IPython.html import widgets
reset_button = widgets.Button(description="Reset")
display(reset_button)

def on_reset_button_clicked(b):
    plot_slice(n=9,dim=2)

reset_button.on_click(on_reset_button_clicked)
```
